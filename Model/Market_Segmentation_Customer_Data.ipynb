```python
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(), cmap="crest",annot=True , fmt=".0%" )
plt.show()
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scaled_df = scalar.fit_transform(df)
```

# Dimensionality reduction
Converting the DataFrame into 2D DataFrame for visualization

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_df)
pca_df = pd.DataFrame(data=principal_components ,columns=["PCA1","PCA2"])
pca_df
```

# Finding Number of Clusters by Elbow Method

```python
from sklearn.cluster import KMeans
```

```python
WCSS = []
range_val = range(1,15)
for i in range_val:
    kmean = KMeans(n_clusters=i, init="k-means++", random_state=0)
    kmean.fit(pd.DataFrame(scaled_df))
    WCSS.append(kmean.inertia_)
    
plt.plot(range_val,WCSS,'bx-')
plt.xlabel('Number of Clusters') 
plt.ylabel('Within Sum of squared error') 
plt.title('Elbow Method') 
plt.show()
```

# KMeans Model

```python
kmeans_model=KMeans(n_clusters=4, init="k-means++", random_state=0)
y_pred=kmeans_model.fit_predict(scaled_df)
y_pred
```

```python
pca_df['cluster']=y_pred
pca_df.head()
```

```python
kmeans_model.cluster_centers_
```

```python
plt.figure(figsize=(8,8))
ax=sns.scatterplot(x="PCA1",y="PCA2",hue="cluster",data=pca_df,palette=['red','green','orange','blue'])
plt.scatter(kmeans_model.cluster_centers_[:,0],kmeans_model.cluster_centers_[:,1],color='black',marker='*',label='centroid')
plt.title("Clustering using K-Means Algorithm")
plt.legend(title="Cluster")
plt.show()
```

```python
cluster_df = pd.concat([df,pd.DataFrame({'Cluster':kmeans_model.labels_})],axis=1)
cluster_df
```

# Saving the kmeans clustering model and the data with cluster label

```python
#Saving Scikitlearn models
import joblib
joblib.dump(kmeans_model, "kmeans_model.pkl")
```

```python
cluster_df.to_csv("Clustered_Customer_Data.csv")
```

# Decision Tree Model

```python
X = cluster_df.drop(['Cluster'],axis=1)
y= cluster_df[['Cluster']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size=0.2,random_state=0)
```

```python
from sklearn.metrics import classification_report
from sklearn import metrics
```

```python
from sklearn.tree import DecisionTreeClassifier
model= DecisionTreeClassifier(criterion="entropy")
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
print(metrics.confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

# Saving the Decision tree model for future prediction

```python
import pickle
filename = 'trained_model.sav'
pickle.dump(model, open(filename, 'wb'))
 
loaded_model = pickle.load(open(filename, 'rb'))
result = loaded_model.score(X_test, y_test)
print('Acuuracy',result)
```
